## Cleaning Data
Data cleaning means fixing bad data in your data set. Bad data could be:
* Empty cells
* Data in wrong format
* Wrong data
* Duplicates

```python
#Read CSV file with electric vehicle information 
df1 = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df1.info()
```

```python
df1.head()
```

Where can see that we have many column with missing values.

### Remove rows

Remove all empty cells and return a clean dataframe.

```python
new_df1 = df1.dropna()
new_df1.info()
```

### Fill empty values with mean, median and mode

Fill specific columns with their mean, median, and mode.

```python
#Let's use the calorie dataframe again.
df.info()
df.head()
```

```python
#Find the mean for Calorines
x = df.Calories.mean()

#Uncomment if you want to use the median or mode
#x = df.Calories.median()
#x = df.Calories.mode()

#Fill the emply values with mean. The inplace parameter modifies the original dataframe
df.Calories.fillna(x, inplace = True)
df.info()
```

### Correcting Wrong Data Format

Cells with data of wrong format can make it difficult, or even impossible, to analyze data. To fix it, you have two options: remove the rows, or convert all cells in the columns into the same format.

```python
df2 = pd.read_csv("Data_with_date.csv")
df2.head(35)
```

Row 26 has the wrong format. Let's fix this with to_datetime().

```python
df2['Date'] = pd.to_datetime(df2['Date'])
df2.head(35)
```

Let's remove the row with the missing value in the date column.

```python
df2.dropna(subset=['Date'], inplace=True)
```

```python
df2.head(35)
```

### Cleaning Wrong Data

Let's clean values that were entered incorrectly. Row 7 looks like it had the duration entered incorrectly.

```python
df2.head(35)
```

```python
#Let's fix row 7.
df2.loc[7, 'Duration'] = 45
df2.head(35)
```

For large datasets you will want to loop through the data within boundaries and replace or drop values that fall outside the guardrails.

```python
#loop through and make sure the max value for duration is 120 minutes
for x in df2.index:
    if df2.loc[x,"Duration"] > 120:
        df2.loc[x,"Duration"] = 120
```

You can also drop the extreme values

```python
for x in df2.index:
    if df2.loc[x,"Duration"] > 120:
        df2.drop(x, inplace = True)
```
